### CODIGO DE EXTRACCION UNO A LA VEZ
- Provincia cadiz, extraeremos 15 puntos, desde la fecha 2016 enero 1 hasta 2026 enero 1. 


```python





import requests
from datetime import datetime, timedelta
import time

# 1. Tu Configuración
API_KEY = "eyJhbGciOiJIUzI1NiJ9.eyJzdWIiOiJIb3JhY2lvdmFsbGVqbzIyQGdtYWlsLmNvbSIsImp0aSI6ImQ2YjE2ODM3LTQxOTEtNDliOS1iMmQ3LTIwYTNkNGVjOTUyNiIsImlzcyI6IkFFTUVUIiwiaWF0IjoxNzY5NDUyODg4LCJ1c2VySWQiOiJkNmIxNjgzNy00MTkxLTQ5YjktYjJkNy0yMGEzZDRlYzk1MjYiLCJyb2xlIjoiIn0.MxpJubjynmtCAQPwF0jvTZNrcdVEBoIV5ee7r50YNGE"
HEADERS = {"api_key": API_KEY}
BASE_URL = "https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos"

# Lista de estaciones (Idemas) que quieres consultar


# Configuración de fechas
fecha_inicio = datetime(2016, 1, 1)
fecha_fin = datetime(2026, 1, 28) 

# 2. Bucle principal: Iterar por cada estación de la lista
for idema in idemas_cadiz:
    print(f"--- Procesando estación: {idema} ---")
    
    current_date = fecha_inicio
    
    # 3. Bucle interno: Iterar día a día usando datetime
    while current_date <= fecha_fin:
        
        # Formatear la fecha al estilo AEMET: "YYYY-MM-DDTHH:MM:SSUTC"
        # Para datos diarios, pedimos inicio y fin el mismo día para obtener solo ese registro
        fecha_str = current_date.strftime("%Y-%m-%dT00:00:00UTC")
        fecha_fin_str = current_date.strftime("%Y-%m-%dT23:59:59UTC")
        
        # Construir el endpoint dinámico
        endpoint = f"{BASE_URL}/fechaini/{fecha_str}/fechafin/{fecha_fin_str}/estacion/{idema}"
        
        try:
            # PRIMERA PETICIÓN: Obtener la URL de los datos
            response = requests.get(url=endpoint, headers=HEADERS)
            
            if response.status_code == 200:
                json_response = response.json()
                
                # Verificar si AEMET devolvió una URL de datos válida
                if json_response.get('estado') == 200 and 'datos' in json_response:
                    url_datos = json_response['datos']
                    
                    # SEGUNDA PETICIÓN: Descargar el JSON real
                    response_datos = requests.get(url_datos)
                    datos_clima = response_datos.json()

                    with open(file = f"cadiz/{idema}_{fecha_str.split("T")[0]}.json", mode = "w", encoding = "utf-8") as file:
                        json.dump(datos_clima, file)
                    
                    # print(f"Datos recibidos para {idema} el {fecha_str[:10]}: {datos_clima}")
                    # Aquí podrías guardar 'datos_clima' en un CSV o Base de Datos
                else:
                    print(f"AEMET no tiene datos para {idema} en {fecha_str[:10]} (Mensaje: {json_response.get('descripcion')})")
            else:
                print(f"Error en la petición inicial: {response.status_code}")
            sleep(1.5)

        except Exception as e:
            print(f"Ocurrió un error: {e}")
        
        # Avanzar un día
        current_date += timedelta(days=1)
        
        # Pausa pequeña para no saturar la API y evitar bloqueos (Rate Limiting)
        time.sleep(1.5)

print("Proceso finalizado.")
```

### UNIFICAR EXTRACCION EN UN SOLO JSON 

```python
import os
import json
import glob
import pandas as pd


carpeta = "cadiz"


archivos = glob.glob(os.path.join(carpeta, "*.json"))

print(f" Encontrados {len(archivos)} archivos JSON")

todos_los_datos = []


for archivo in archivos:
    with open(archivo, "r", encoding="utf-8") as f:
        data = json.load(f)

        # Si es lista → extender
        if isinstance(data, list):
            todos_los_datos.extend(data)

        # Si es dict → añadir como registro
        elif isinstance(data, dict):
            todos_los_datos.append(data)

print(f" Registros totales combinados: {len(todos_los_datos)}")


df = pd.DataFrame(todos_los_datos)

print("\n DataFrame preview:")
display(df.head())

salida = "cadiz_unificado.json"

with open(salida, "w", encoding="utf-8") as f:
    json.dump(todos_los_datos, f, ensure_ascii=False, indent=2)

print(f"\n  Archivo generado: {salida}")

```
